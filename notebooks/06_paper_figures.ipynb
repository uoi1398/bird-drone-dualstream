```python
import os
import pandas as pd
import matplotlib.pyplot as plt

os.makedirs("outputs/figures", exist_ok=True)

main_results = pd.read_csv("outputs/tables/main_results.csv")
display(main_results)
```

```python
plot_df = main_results.copy()

plot_df["test_accuracy_percent"] = plot_df["test_accuracy"] * 100
plot_df["test_macro_f1_percent"] = plot_df["test_macro_f1"] * 100

display(plot_df)
```

```python
import numpy as np

conditions = plot_df["condition"].unique()
models = plot_df["model"].unique()

x = np.arange(len(conditions))
width = 0.25

plt.figure(figsize=(8, 5))

for i, model_name in enumerate(models):
    values = []
    for cond in conditions:
        value = plot_df[
            (plot_df["condition"] == cond) &
            (plot_df["model"] == model_name)
        ]["test_accuracy_percent"].values[0]
        values.append(value)

    plt.bar(x + i * width, values, width, label=model_name)

plt.xticks(x + width, conditions, rotation=20, ha="right")
plt.ylabel("Accuracy (%)")
plt.title("Test Accuracy under Different Visual Conditions")
plt.legend()
plt.tight_layout()

plt.savefig("outputs/figures/test_accuracy_bar.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8, 5))

for i, model_name in enumerate(models):
    values = []
    for cond in conditions:
        value = plot_df[
            (plot_df["condition"] == cond) &
            (plot_df["model"] == model_name)
        ]["test_macro_f1_percent"].values[0]
        values.append(value)

    plt.bar(x + i * width, values, width, label=model_name)

plt.xticks(x + width, conditions, rotation=20, ha="right")
plt.ylabel("Macro-F1 (%)")
plt.title("Test Macro-F1 under Different Visual Conditions")
plt.legend()
plt.tight_layout()

plt.savefig("outputs/figures/test_macro_f1_bar.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
paper_table = plot_df.pivot_table(
    index=["condition", "selected_alpha"],
    columns="model",
    values=["test_accuracy_percent", "test_macro_f1_percent"]
)

display(paper_table)
```

```python
paper_table_path = "outputs/tables/paper_main_results_table.csv"
paper_table.to_csv(paper_table_path)

print("Saved to:", paper_table_path)
```
