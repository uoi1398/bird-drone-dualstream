```python
from pathlib import Path

DATA_ROOT = Path("data/raw")

VIDEO_EXTS = [".mp4", ".avi", ".mov", ".mkv"]

def collect_all_videos(root, video_exts):
    videos = []
    for ext in video_exts:
        videos.extend(root.rglob(f"*{ext}"))
    return sorted(videos)

all_videos = collect_all_videos(DATA_ROOT, VIDEO_EXTS)

print("All videos:", len(all_videos))

print("\nFirst 20 videos:")
for p in all_videos[:20]:
    print(p)
```

```python
visible_videos = [
    p for p in all_videos
    if "VIDEO_V" in str(p).upper()
]

print("Visible videos:", len(visible_videos))

print("\nFirst 20 visible videos:")
for p in visible_videos[:20]:
    print(p)
```

```python
bird_videos = [
    p for p in visible_videos
    if "BIRD" in p.name.upper()
]

drone_videos = [
    p for p in visible_videos
    if "DRONE" in p.name.upper()
]

print("Bird visible videos:", len(bird_videos))
print("Drone visible videos:", len(drone_videos))

print("\nExample bird videos:")
for p in bird_videos[:5]:
    print(p)

print("\nExample drone videos:")
for p in drone_videos[:5]:
    print(p)
```

```python
assert len(bird_videos) == 51, f"Expected 51 bird videos, got {len(bird_videos)}"
assert len(drone_videos) == 114, f"Expected 114 drone videos, got {len(drone_videos)}"

print("Video counts are correct.")
```

```python
import random
import pandas as pd

SEED = 42
random.seed(SEED)

bird_videos_shuffled = bird_videos.copy()
drone_videos_shuffled = drone_videos.copy()

random.shuffle(bird_videos_shuffled)
random.shuffle(drone_videos_shuffled)

records = []

def add_records(video_paths, label, label_name, train_n, val_n, test_n):
    assert len(video_paths) == train_n + val_n + test_n

    split_names = (
        ["train"] * train_n +
        ["val"] * val_n +
        ["test"] * test_n
    )

    for path, split in zip(video_paths, split_names):
        records.append({
            "video_id": path.stem,
            "path": str(path),
            "label": label,
            "label_name": label_name,
            "split": split
        })

# bird = 0, drone = 1
add_records(
    bird_videos_shuffled,
    label=0,
    label_name="bird",
    train_n=36,
    val_n=7,
    test_n=8
)

add_records(
    drone_videos_shuffled,
    label=1,
    label_name="drone",
    train_n=79,
    val_n=18,
    test_n=17
)

manifest = pd.DataFrame(records)

display(manifest.head())

print("Total samples:", len(manifest))

print("\nSplit x label:")
print(pd.crosstab(manifest["split"], manifest["label_name"]))
```
