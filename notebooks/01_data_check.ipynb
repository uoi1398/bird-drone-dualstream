```python
import sys
sys.path.append("src")

import os
import pandas as pd
import torch

from dataset import BirdDroneVideoDataset
```

```python
CSV_PATH = "data/splits/manifest.csv"

df = pd.read_csv(CSV_PATH)

print(df.head())
print(df.columns)
print("Total samples:", len(df))
```

```python
print("Label counts:")
print(df["label_name"].value_counts())

print("\nSplit counts:")
print(df["split"].value_counts())

print("\nSplit x Label:")
print(pd.crosstab(df["split"], df["label_name"]))
```

```python
df["path_exists"] = df["path"].apply(os.path.exists)

print(df["path_exists"].value_counts())

missing = df[~df["path_exists"]]

if len(missing) > 0:
    print("Missing files:")
    display(missing.head(20))
else:
    print("All video paths exist.")
```

```python
rgb_ds = BirdDroneVideoDataset(
    csv_path=CSV_PATH,
    split="train",
    mode="rgb",
    num_frames=16,
    image_size=224,
    degrade_level="none"
)

x, y, video_id = rgb_ds[0]

print("RGB tensor shape:", x.shape)
print("Label:", y)
print("Video ID:", video_id)
```

```python
flow_ds = BirdDroneVideoDataset(
    csv_path=CSV_PATH,
    split="train",
    mode="flow",
    num_frames=8,
    image_size=160,
    degrade_level="none"
)

x_flow, y_flow, video_id_flow = flow_ds[0]

print("Flow tensor shape:", x_flow.shape)
print("Label:", y_flow)
print("Video ID:", video_id_flow)
```
