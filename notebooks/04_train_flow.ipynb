```python
import pandas as pd

log_path = "runs/20260515_flow_seed42/train_log.csv"
log_df = pd.read_csv(log_path)

display(log_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(log_df["epoch"], log_df["train_macro_f1"], marker="o", label="Train Macro-F1")
plt.plot(log_df["epoch"], log_df["val_macro_f1"], marker="o", label="Val Macro-F1")
plt.xlabel("Epoch")
plt.ylabel("Macro-F1")
plt.title("Flow-only Training Curve")
plt.legend()
plt.grid(True)
plt.show()
```
